```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
values = [0.9534711184772267, 0.9469786240504339, 0.9330407185801144, 0.971619642362362, 0.9580025249405911, 0.9739887807104323, 0.9439404408137003, 0.9619000113980715]
labels = ['USD', 'JPY', 'GOLD', 'AUD', 'NZD', 'GBP', 'EUR', 'CAD']

values_normalized = np.array(values)
values_normalized = (values_normalized - np.min(values_normalized)) / np.ptp(values_normalized)

plt.barh(labels, values_normalized)
plt.xlabel('Normalized Values')
plt.ylabel('Labels')
plt.title('Normalized Bar chart')

plt.show()
```
